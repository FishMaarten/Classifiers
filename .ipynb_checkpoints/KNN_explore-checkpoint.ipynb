```python
import time
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
from sklearn import metrics
from sklearn import datasets
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
```

# Score different preset parameters

```python
score = {
    "n_neighbors": {neighbors:[] for neighbors in [1,2,3,5,10,20,50,100]},
    "leaf_size": {leaf:[] for leaf in range(20,40 +1)},
    "weights": {weights:[] for weights in ["uniform","distance"]},
    "p": {distance:[] for distance in range(1,5 +1)}
}
```

```python
count = 2 # 10 seconds per count
start = time.time()
for kwarg, data in score.items():
    for val in score[kwarg].keys():
        for idx in range(count):
            X_train, X_test, y_train, y_test = train_test_split(
                digits.data, digits.target, test_size=0.3)
            clf = KNeighborsClassifier(**{kwarg : val}).fit(X_train, y_train)
            score[kwarg][val].append(metrics.accuracy_score(y_test, clf.predict(X_test)))
print(f"Calculated {count} iterations in :", round(time.time() -start, 2), "seconds")
```

# Plot different parameter influence

```python
for classifier in score:
    stats = pd.DataFrame(score[classifier]).describe().drop(["count","std"]).T
    stats.plot.bar(title=classifier, figsize=(16,4), grid=True,
                   ylim=(round(stats.min().min() -0.01, 2),1))
    plt.show()
```

# Test different parameter options

```python
params = {
    "min": {k: pd.DataFrame(score[k]).min().idxmin() for k in score.keys()},
    "avg": {k: pd.DataFrame(score[k]).mean().idxmax() for k in score.keys()},
    "std": {k: pd.DataFrame(score[k]).std().idxmin() for k in score.keys()},
    "max": {k: pd.DataFrame(score[k]).max().idxmax() for k in score.keys()}
}
```

```python
evaluation = {k:[] for k in params.keys()}
for idx in range(10):
    for key, kwargs in params.items():
        X_train, X_test, y_train, y_test = train_test_split(
            digits.data, digits.target, test_size=0.3)
        clf = KNeighborsClassifier(**kwargs).fit(X_train, y_train)
        evaluation[key].append(metrics.accuracy_score(y_test, clf.predict(X_test)))
```

```python
df = pd.DataFrame(evaluation).describe().drop(["count","std","mean"]).T
df.plot.bar(figsize=(16,4), grid=True, ylim=(round(df.min().min() -0.01, 2), 1)); plt.show()
```

# Best performing parameters

```python
arr = []
floor = 0.98
start = time.time()
for idx in range(100):
    X_train, X_test, y_train, y_test = train_test_split(
        digits.data, digits.target, test_size=0.3)
    clf = KNeighborsClassifier(**params["max"]).fit(X_train, y_train)
    arr.append(metrics.accuracy_score(y_test, clf.predict(X_test)))
print("Calculated in :", round(time.time() -start, 2), "seconds")
print(f"Number of low scores < {int(floor*100)}% :",
      pd.DataFrame(arr)[0].apply(lambda x: x < floor).sum())
```
