```python
import time
import json
import numpy as np
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

from sklearn import metrics
from sklearn import datasets

from sklearn.svm import SVC
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier

from sklearn.model_selection import train_test_split

digits = datasets.load_digits()
```

# Score different preset parameters

```python
classifiers = { 
    # Holds the classifier objects
    "svc"  : SVC,
    "knn"  : KNeighborsClassifier,
    "tree" : DecisionTreeClassifier,
    "nb"   : GaussianNB}
titles = { 
    # Classifier string for plot title
    "svc"  : "Support Vector Machine",
    "knn"  : "K-Nearest Neighbors",
    "tree" : "Decision Tree",
    "nb"   : "Gaussian Naive Bayes"}

if False: # Builds dictionary from scratch
    # Dict entry for each impactful parameter-values of different classifiers
    # properties[clf][param]         = {value} each impactful parameter value
    # properties[clf][param][preset]  = [score] of parameter accuracy
    properties = {   
        "svc": {  "kernel"       : {kernel:    [] for kernel in ["linear", "poly", "rbf"]},
                  "gamma"        : {gamma:     [] for gamma in ["auto", "scale"]},
                  "C"            : {C:         [] for C in [1,2,3,5,10]}},
        
        "knn": {  "n_neighbors"  : {neighbors: [] for neighbors in [1,2,3,5,10]},
                  "leaf_size"    : {leaf:      [] for leaf in range(21,40,3)},
                  "weights"      : {weights:   [] for weights in ["uniform","distance"]},
                  "p"            : {distance:  [] for distance in range(1,5 +2)}},
        
        "tree": { "criterion"    : {crit:      [] for crit in ["gini","entropy"]},
                  "splitter"     : {split:     [] for split in ["best","random"]},
                  "max_depth"    : {depth:     [] for depth in [2,3,5,8,10]},        
                  "max_features" : {feat:      [] for feat in [0.1,0.5,0.9,1,2,5,"auto","sqrt","log2"]}},
        
        "nb": {   "var_smoothing": {smooth:    [] for smooth in [1e-12,1e-9,1e-7]}}}

    count = 5 # takes up to 10 seconds per count
    total = time.time()
    for clf in classifiers:
        for param, score in properties[clf].items():
        # Loop through each classifier's parameters
            start = time.time()
            for idx in range(count):
                # Iterate count-times, use same data split for more accurate metrics
                X_train, X_test, y_train, y_test = train_test_split(
                    digits.data, digits.target, test_size=0.3)   
                for val in score.keys():
                    # Loop through each impactful parameter value & append accuracy
                    model = classifiers[clf](**{param: val}).fit(X_train, y_train)
                    properties[clf][param][val].append(metrics.accuracy_score(y_test, model.predict(X_test)))
            print(f"Calculated {clf}[{param}] :", round(time.time() -start, 2), "seconds")
    print(f"Total time for {count} iterations :", round(time.time() -total, 2), "seconds")
    
else: # Load pre-calculated properties dictionary from JSON file
    def is_float(val):
        # Return true if val is castable to type float
        try: float(val);   return True
        except ValueError: return False
    with open('class_metrics.json', 'r', encoding='utf-8') as data:
        properties = json.load(data, parse_float=np.float64)
        for clf in classifiers:
            for param, values in properties[clf].items():
                # Copy properties with comprehension and cast to int|float 
                properties[clf][param] = {
                    int(k) if k.isnumeric() else float(k) if is_float(k) else k:v 
                    for k,v in properties[clf][param].items()} 
```

# Plot different parameter influence

```python
GaussianNB()
```

```python
dfs = {} # Holds all the dataframes for each class' parameter statistics
for clf in properties.keys():
    dfs[clf] = {}
    size = len(properties[clf])
    fig, axes = plt.subplots(nrows=size, ncols=1, figsize=(12,3*size))
    fig.suptitle(titles[clf], fontsize=30, x=0.33, y=.9, va="baseline")
    for idx, (key, kwarg) in enumerate(properties[clf].items()):
       # Create dataframes for each classifiers' parameters, store in dfs dictionary
        dfs[clf][key] = pd.DataFrame(kwarg).describe().drop(["count","std"]).T
        stats = dfs[clf][key]
        stats.plot.bar(ax=axes[idx] if type(axes)==np.ndarray else axes, legend=False,
            rot=0, grid=True, ylim=(round(stats.min().min() - 0.05, 2),1),
            color=["lightblue","#ffa600","#ff6361","#bc5090","#58508d","#003f5c"])
        plt.text(-0.1, 0.1, key, rotation=90, fontsize=20, 
            transform=axes[idx].transAxes if type(axes)==np.ndarray else axes.transAxes)
    # Plotting all parameter statistics
    plt.show()
```

# Best performing parameters

```python
compositions = {clf: { # Dict entry per classifier, threshold parameters
# Loads dataframe for current clf-prop and sets preset kwargs 
    "min": {k: dfs[clf][k]["min"].idxmax() for k in dfs[clf]}, # Highest low-score value
    "avg": {k: dfs[clf][k]["50%"].idxmax() for k in dfs[clf]}, # Highest average value
    "qrt": {k: dfs[clf][k]["75%"].idxmax() for k in dfs[clf]}, # Highest 75% value
    "max": {k: dfs[clf][k]["max"].idxmax() for k in dfs[clf]}, # Highest max value
}   for clf in classifiers.keys()}

performance = {}
for clf in classifiers:
    start = time.time()
    # Initialize {composition: [accuracy]}
    performance[clf] = {k:[] for k in compositions[clf].keys()}
    # Set floor to break-point accuracy of the current classifier
    floor = np.array([dfs[clf][k]["25%"].max() for k in dfs[clf]]).min()
    #for comp, kwargs in compositions[clf].items():
    for comp, kwargs in compositions[clf].items():
        # Loop through each classifiers' compositions
        for idx in range(10):
            # Build 10 models for each composition and append accuracy
            X_train, X_test, y_train, y_test = train_test_split(
                digits.data, digits.target, test_size=0.3)
            model = classifiers[clf](**kwargs).fit(X_train, y_train)
            performance[clf][comp].append(metrics.accuracy_score(y_test, model.predict(X_test)))
        # Output composition kwargs, calculation time & number of low scores
        print(compositions[clf][comp])
        print(f"Calculated {clf}[{comp}] in :", round(time.time() -start, 2), "seconds")
        print(f"Number of low scores < {round(floor*100, 2)}% :",
            pd.DataFrame(performance[clf][comp])[0].apply(lambda x: x < floor).sum(), "\n")
    print("\n")
```
